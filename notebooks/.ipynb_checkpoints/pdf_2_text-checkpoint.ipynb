```python
import PyPDF2
```

```python
paper="A National Trauma Care System- Integrating Military and Civilian Trauma Systems to Achieve Zero Preventable Deaths After Injury.pdf"
paper="Context dependency of medical devices.pdf"
paper='Copycats? Digital consumers in the online age.pdf'
pdfFileObj=open('papers/'+paper,"rb")
```

```python
pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
pdfReader.numPages
```

```python
all_pages=[]
```

```python
for page in range(pdfReader.numPages):
    all_pages.append(pdfReader.getPage(page).extractText())
```

```python
all_pages_clean=[page.replace(' ','') for page in all_pages]
all_pages_clean=[page.replace('\t','') for page in all_pages_clean]
all_pages_clean=[page.replace('\n','') for page in all_pages_clean]
all_pages_clean=[page.replace('-','') for page in all_pages_clean]
all_pages_clean=[page.lower() for page in all_pages_clean]
```
