```python
import json
import csv
import requests
import io
from bs4 import BeautifulSoup
from IPython.core.display import display, HTML
import os

crossref_folder='crossref'
journal='jmis'
pages=2
alt_base='https://api.altmetric.com/v1/doi/'
dois=[]
for file_index in range(pages):
    json_file='../'+crossref_folder+'/'+journal+str(file_index)+'.json'
    with open(json_file, 'r') as f:
        json_data = json.load(f)
        dois_per_file=[item['DOI'] for item in json_data['message']['items']]
        dois=dois+dois_per_file
len(dois)
```

```python
alt={}
for doi in dois:
    alt_url=alt_base+doi
    r = requests.get(alt_url)
    #print(r.status_code, end=',')
    print (r.status_code,end=',')
    if r.status_code == 200:
        alt[doi]=r.json()
```

```python
with open(journal+'_alt.json', 'w') as outfile:
    json.dump(alt, outfile)
```
